```python
# Project Title
# Analyzing and Visualizing Movie Ratings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating Movie Dataset
data = {
    'Movie': ['Inception', 'Avatar', 'Titanic', 'Interstellar', 'Joker',
              'Avengers', 'Frozen', 'Bahubali', 'RRR', 'KGF'],

    'Genre': ['Sci-Fi', 'Action', 'Romance', 'Sci-Fi', 'Drama',
              'Action', 'Animation', 'Action', 'Action', 'Action'],

    'Rating': [8.8, 7.9, 7.8, 8.6, 8.5,
               8.0, 7.5, 8.2, 8.4, 8.3]
}

# Convert into DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Movie Dataset")
print(df)
```

```python
# Check Missing Values
print("\nMissing Values")
print(df.isnull().sum())
```

```python
# Summary Statistics
print("\nSummary Statistics")
print("Mean Rating:", df['Rating'].mean())
print("Median Rating:", df['Rating'].median())
print("Mode Rating:", df['Rating'].mode()[0])
```

```python
# Histogram
plt.figure(figsize=(8,5))
plt.hist(df['Rating'])
plt.title("Distribution of Movie Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Rating'])
plt.title("Boxplot of Ratings")
plt.show()
```

```python
# Top Rated Movies
print("\nTop Rated Movies")
top_movies = df.sort_values(by='Rating', ascending=False)
print(top_movies)
```

```python
# Genre Distribution Plot
plt.figure(figsize=(8,5))
sns.countplot(x='Genre', data=df)
plt.title("Genre Distribution")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
