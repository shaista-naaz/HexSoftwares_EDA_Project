```python
from sklearn.datasets import load_iris
import pandas as pd

# Load dataset
iris = load_iris()

# Convert dataset into dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Display first 5 rows
print(df.head())
```

```python
# Dataset information
print(df.info())

# Statistical summary
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Histogram graph
df.hist(figsize=(10,8))

plt.show()
```

# Conclusion

- The Iris dataset contains 150 flower samples.
- The dataset has 4 numerical features.
- Petal length and petal width show clear variations.
- Histograms helped visualize data distribution.
- The dataset is clean and suitable for analysis.
